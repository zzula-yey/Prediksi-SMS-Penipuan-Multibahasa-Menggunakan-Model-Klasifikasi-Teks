```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

# ========================================================
from NaiveBayesManual import NaiveBayesSimple
```

<a id="1"></a>
# <p style="background-color:#034C53;font-family:newtimeroman;color:#FFF9ED;font-size:130%;text-align:center;border-radius:10px 10px;">DATASET UNDERSAMPLING</p>

<a id="1"></a>
## <p style="background-color:#034C53;font-family:newtimeroman;color:#FFF9ED;font-size:110%;text-align:center;border-radius:10px 10px;">BACA DATASET</p>

```python
data = pd.read_csv('./dataset (undersampling-oversampling)/dataset_sms_undersampling.csv')
data
```

<a id="1"></a>
## <p style="background-color:#034C53;font-family:newtimeroman;color:#FFF9ED;font-size:110%;text-align:center;border-radius:10px 10px;">FEATURE ENGINEERING</p>

```python
x = data['clean_Message']
y = data['label']
```

**Pakai TF-IDF dan N-Gram**

```python
vec_TF_IDF = TfidfVectorizer(ngram_range=(1,1))
vec_TF_IDF.fit(x)

# x_tf_idf = vec_TF_IDF.transform(x)
```

```python
X = vec_TF_IDF.transform(x).toarray()
X
```

```python
data_tabular_tf_idf = pd.DataFrame(X,columns=vec_TF_IDF.get_feature_names_out())
data_tabular_tf_idf
```

```python
y = np.array(y)
y
```

**Simpan menjadi Pickle**

```python
pickle.dump(vec_TF_IDF.vocabulary_,open("./file_pickle/feature_tf-idf_undersampling.sav", "wb"))
```

<a id="1"></a>
## <p style="background-color:#034C53;font-family:newtimeroman;color:#FFF9ED;font-size:110%;text-align:center;border-radius:10px 10px;">MODEL</p>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

nb = NaiveBayesSimple()
model = nb.train(X_train, y_train)

y_pred = model.predict(X_test)

print("Akurasi:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Simpan model ke dalam file
pickle.dump(model,open("./file_pickle/model_fraud_undersampling.sav","wb" ))
```
